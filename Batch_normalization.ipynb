```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.2, random_state=42)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Complex Synthetic Classification Dataset')
plt.show()
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization

model=Sequential()
model.add(Dense(2,activation='relu',input_dim=2))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history1 = model.fit(X,y,epochs=200,validation_split=0.2,verbose=0)
```

```python
plt.plot(history1.history['val_loss'],color='yellow')
plt.plot(history1.history['val_accuracy'],color='green')
# plotting val_accuracy and val loss
```

```python
# making another model with batch nomalization

model1=Sequential()

model1.add(Dense(2,activation='relu',input_dim=2))
model1.add(BatchNormalization())
model1.add(Dense(2,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(1,activation='sigmoid'))

model1.summary()
```

in the above model summary we can see that bacth normalization have 8 parameters in which 4 parameters are trainable and 4 are non trainable

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history2=model1.fit(X,y,epochs=200,validation_split=0.2,verbose=0)
```

```python
plt.plot(history2.history['val_accuracy'],color='green')
plt.plot(history2.history['val_loss'],color='black')
```

```python
plt.plot(history1.history['val_accuracy'],color='green')# history1 =model without batch normalization
plt.plot(history2.history['val_accuracy'],color='red')# history2 = model with  batch normalization
```

in the above figure we can see that the val_accuracy has been increase just by using batch normalization

```python
plt.plot(history1.history['val_loss'],color='yellow')
plt.plot(history2.history['val_loss'],color='black')
```

here we can see that how much difference in val_loss is there just by using Batch normalization
