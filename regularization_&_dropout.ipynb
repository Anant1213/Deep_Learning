we are making a classification data and first we are trying normal model training and testing it .
in the second part we are using regularization tehcnique L2
in the 3rd part we are using dropout layer for smoothening

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
X,y=make_moons(100,noise=0.25,random_state=2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
model1 = Sequential()
model1.add(Dense(128, input_dim=2, activation='relu' ))
model1.add(Dense(128, activation='relu'))
model1.add(Dense(1, activation='sigmoid'))

model1.summary()
```

```python
adam=Adam(learning_rate=0.01)
model1.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])

history1=model1.fit(X,y,epochs=2000,validation_split=0.2,verbose=0)
```

```python
plot_decision_regions(X,y.astype('int'),clf=model1,legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

Here we are trying regularization method for model12

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model12 = Sequential()
model12.add(Dense(128, input_dim=2, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.03) ))
model12.add(Dense(128, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.03)))
model12.add(Dense(1, activation='sigmoid'))

model12.summary()
```

```python
adam=Adam(learning_rate=0.01)
model12.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])

history12=model12.fit(X,y,epochs=2000,validation_split=0.2,verbose=0)
```

```python
plot_decision_regions(X,y.astype('int'),clf=model12,legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
# as we can see it has worked better and decreased the problem of overfitting
```

```python
plt.plot(history12.history['loss'])
plt.plot(history12.history['val_loss'])
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

model123 = Sequential()
model123.add(Dense(128, input_dim=2, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.03) ))
model123.add(Dropout(0.3))
model123.add(Dense(128, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.03)))
model123.add(Dropout(0.3))
model123.add(Dense(1, activation='sigmoid'))

model123.summary()
```

```python
adam=Adam(learning_rate=0.01)
model123.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])

history123=model123.fit(X,y,epochs=2000,validation_split=0.2,verbose=0)
```

```python
plot_decision_regions(X,y.astype('int'),clf=model123,legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
#here we can see we have made a little buit more change in overfitting by using dropouty layer
```

```python
plt.plot(history123.history['loss'])
plt.plot(history123.history['val_loss'])
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

model1234 = Sequential()
model1234.add(Dense(128, input_dim=2, activation='relu',kernel_regularizer=tf.keras.regularizers.l1(0.003) ))
model1234.add(Dropout(0.3))
model1234.add(Dense(128, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model1234.add(Dropout(0.3))
model1234.add(Dense(1, activation='sigmoid'))

model1234.summary()
adam=Adam(learning_rate=0.01)
model1234.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])

history1234=model1234.fit(X,y,epochs=2000,validation_split=0.2,verbose=0)
plot_decision_regions(X,y.astype('int'),clf=model1234,legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
#here we can see we have made  change in overfitting by using dropouty layer and l1 regression adn l2 regressioin for respective layer
```

```python
plt.plot(history1234.history['loss'])
plt.plot(history1234.history['val_loss'])
```
